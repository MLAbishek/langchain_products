```python
import os

from openai import OpenAI

client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key=os.getenv("OPENROUTER_API_KEY"),
)

response = client.chat.completions.create(
  model="openai/gpt-oss-120b:free",
  messages=[
          {
            "role": "user",
            "content": "How many r's are in the word 'strawberry'?"
          }
        ]
)

print(response.choices[0].message)
```

```python
response.choices[0].message.content
```

# **Messages**

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
```

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

from langchain_openai import ChatOpenAI

user_input = input("Enter the Question: ")

messages = [
    SystemMessage(content="You are a Gen-Z assistant who always answers in a fun way"),
    HumanMessage(content=user_input)  
]

llm = ChatOpenAI(
    model="openai/gpt-oss-120b:free",
    base_url="https://openrouter.ai/api/v1",
    api_key=os.getenv("OPENROUTER_API_KEY"),
    temperature=0.7
)

llm.invoke(messages)
```

# Prompts

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate.from_template("What is {user_input}")

final_promt = prompt.invoke(input="Data Science")

llm.invoke(final_promt)
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a Gen-Z Assistant."),
        ("user", "Tell me fun fact about {input1}, then tell me a joke about {input2}")
    ]
)

final_prompt = prompt.invoke({"input1": "Data", "input2": "Science"})


response = llm.invoke(final_prompt)
```

```python
print(response.content)
```

```python
prompt.invoke("Machine Learning")
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {tone} assistant"),
    ("user", "Write a fun fact about {topic}. Give me the response in key-value pair format. Key should be 'topic' and value should be 'fact'.")
])

user_input = input("Enter a topic: ")
user_tone = input("Enter a tone: ")

ready_prompt = prompt_template.invoke({"topic":user_input, "tone":user_tone})

response = llm.invoke(ready_prompt)
print(response)
```

```python
print(response.content)
```

# **Pydantic**

```python
from pydantic import BaseModel

class llm_schema(BaseModel):
    topic: str
    fact: str
```

```python
data = llm_schema(**{"topic":"AI", "fact":"AI is the future of technology"})
print(data)
```

```python
from pydantic import BaseModel

class llm_schema(BaseModel):
    topic: str
    fact: str

llm = ChatOpenAI(
    model="mistralai/mistral-7b-instruct",
    base_url="https://openrouter.ai/api/v1",
    api_key=os.getenv("OPENROUTER_API_KEY"),
    temperature=0.7
)
new_llm = llm.with_structured_output(llm_schema)

response = new_llm.invoke("Write a fun fact about AI.")
response
```
